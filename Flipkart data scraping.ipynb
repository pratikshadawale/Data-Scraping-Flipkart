```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd

```

```python
url = 'https://www.flipkart.com/search?q=mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=mobiles%7CMobiles&requestId=1ea7ca36-2784-4ae6-9f49-c2f8995b9dea&as-searchtext=mobiles'
```

```python
req = requests.get(url)
```

```python
content = BeautifulSoup(req.content,'html.parser')
```

```python
print(content)
```

```python
data = content.find_all('div',{'class':'_13oc-S'}) 
```

```python
print(data)
```

```python
products=[]
prices=[]
memory = []
display = []                  
camera = []                 
battery = []  
ratings=[]
```

```python
for item in data:
    rest_link = item.find('a')['href']
    product = item.find('div',attrs={'class':'_4rR01T'})
    price = item.find('div',attrs={'class':'_30jeq3 _1_WHN1'})
    rating = item.find('div',attrs={'class':'_3LWZlK'})
    specification = item.find('div', attrs={'class':'fMghEO'})
    
    for each in specification:
            col=each.find_all('li', attrs={'class':'rgWa7D'})
            memory_ =col[0].text
            display_ = col[1].text
            camera_ = col[2].text
            battery_ = col[3].text
        
    products.append(product.text)
    prices.append(price.text[1:])
    ratings.append(rating.text)
    memory.append(memory_)
    display.append(display_) 
    camera.append(camera_) 
    battery.append(battery_)
  
```

```python
df = pd.DataFrame({'Product Name':products,'RAM / Internal Memory':memory,
                   'Battery':battery,'Display':display,"Camera":camera,
                   'Prices':prices, 'Ratings':ratings})
```
